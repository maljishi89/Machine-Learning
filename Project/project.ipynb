```python
import sklearn
import pandas as pd
import numpy as np
import sys
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
import sklearn.model_selection
import timeit
from sklearn import datasets
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import cross_validate
from sklearn.model_selection import LeaveOneOut
from sklearn import svm
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score, adjusted_rand_score, confusion_matrix
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, LinearSVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
def makeTest(x_data, y_data):
    scoring2 = {'accuracy': 'accuracy',
               'precision': make_scorer(precision_score, average='weighted'),
               'recall': make_scorer(recall_score, average='weighted'),
               'f1': make_scorer(f1_score, average='weighted'),
               # entropy loss
               'log_loss': 'neg_log_loss',
               # ARI
               'adjusted_rand_score': make_scorer(adjusted_rand_score),
               }

    models = []
    # using 8 differnet models
    models.append(("LogisticRegression",LogisticRegression(solver='lbfgs', max_iter=10000)))
    models.append(("SVC",svm.SVC(kernel='linear', C=1, probability=True)))
    models.append(("Gaussian Naive Bayes",GaussianNB()))
    models.append(("LCA",LinearDiscriminantAnalysis()))
    models.append(("KNeighbors",KNeighborsClassifier(n_neighbors=10)))
    models.append(("DecisionTree",DecisionTreeClassifier()))
    models.append(("RandomForest",RandomForestClassifier()))
    models.append(("AdaBoost",AdaBoostClassifier()))
    models.append(("Bagging",BaggingClassifier()))
    models.append(("Linear Discriminant Analysis",LinearDiscriminantAnalysis()))

#     rf2 = RandomForestClassifier(n_estimators=100, criterion='gini',
#                                     max_depth=10, random_state=0, max_features=None)
#     models.append(("RandomForest2",rf2))


    results = []
    names = []

    for name,model in models:

        model = model
        scoring2 = {'accuracy': 'accuracy',
                   'precision': make_scorer(precision_score, average='weighted'),
                   'recall': make_scorer(recall_score, average='weighted'),
                   'f1': make_scorer(f1_score, average='weighted'),
                   # entropy loss
                   'log_loss': 'neg_log_loss',
                   # ARI
                   'adjusted_rand_score': make_scorer(adjusted_rand_score),
                   }

        start = timeit.default_timer()
        # removed scoring!?
        fold_5 = cross_validate(estimator = model, X=x_data, y = y_data.values.ravel(), cv=5, scoring= scoring2)

        stop = timeit.default_timer()
        print(name, " classifier with 5 fold cross validation")
        print("--------------------------------------------------------------")
        print('Training Time: ', stop - start, 's')  
        names.append(name)
        results.append(fold_5)
        for key, value in fold_5.items():                
            print(key, ' : ', value)

        print("\n==================================================================\n")

    print("**************************Mean of Test accuracy**************************\n")

    for i in range(len(names)):
        me= names[i],results[i]['test_accuracy'].mean()
        print(me[0] + ": " + " {:.3f}".format(me[1]*100) + " %");

    print("\n**************************Mean of precision**************************\n")

    for i in range(len(names)):
        me= names[i],results[i]['test_precision'].mean()
        print(me[0] + ": " + " {:.3f}".format(me[1]));

    print("\n**************************Mean of f1**************************\n")

    for i in range(len(names)):
        me= names[i],results[i]['test_f1'].mean()
        print(me[0] + ": " + " {:.3f}".format(me[1]));

    print("\n**************************Mean of entropy loss**************************\n")

    for i in range(len(names)):
        me= names[i],results[i]['test_log_loss'].mean()
        print(me[0] + ": " + " {:.3f}".format(me[1]));

    print("\n**************************Mean of ARI**************************\n")

    for i in range(len(names)):
        me= names[i],results[i]['test_adjusted_rand_score'].mean()
        print(me[0] + ": " + " {:.3f}".format(me[1]));
```
